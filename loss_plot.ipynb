```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
dict_sovlers = {
  '-3': 'Quasi-Newton',
  '-2': 'Quasi-Newton-learned',
  '-1': 'Deep Quasi-Newton (MoDL)',
   '0': 'Deep ADMM',
   '1': 'ADMM-learned',
   '2': 'ADMM'
}

mpl.style.use('seaborn')
fig, ax = plt.subplots(figsize=(10, 6))

for flag_solver in range(-3, 3):
    File_object = open('logs_plot/Solver='+str(flag_solver)+'_K=1_flag_ND=3_ratio=0.1.log', 'r')
    Loss_string = File_object.readlines()
    Loss_val = []
    for i in range(1, len(Loss_string)//2-1):
        loss_line = Loss_string[i*2]
        Loss_val.append(float(loss_line[-1-7:-1]))
    ax.plot(Loss_val, 'C'+str(flag_solver+3), label=dict_sovlers[str(flag_solver)], linewidth=1.5)
    ax.legend()
ax.set_title('Validation loss curve', color='C0')
plt.xlabel('Epochs')
plt.ylabel('L1 loss')

plt.savefig('plot.png', dpi=600)
```
